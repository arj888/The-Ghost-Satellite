```python
# 1. Sabse pehle astronomy library install karenge (Yeh sirf 5-10 second legi)
!pip install skyfield requests -q

import requests
from skyfield.api import Topos, load

# 2. CelesTrak se live active satellites ka TLE data download karna (Cloud-to-Cloud, 0MB Mobile Data)
# Hum testing ke liye 'Visual' satellites ki list utha rahe hain jisme sabse bade satellites hote hain
tle_url = "https://celestrak.org/NORAD/elements/gp.php?GROUP=visual&FORMAT=tle"
print("=== THE GHOST SATELLITE: EXTRACTION ACTIVE ===")
print("Fetching real-time TLE database from CelesTrak...")

response = requests.get(tle_url)
tle_data = response.text.splitlines()

print(f"Database Loaded successfully. Total raw data lines: {len(tle_data)}\n")

# 3. Skyfield ka use karke orbital calculation engine setup karna
ts = load.timescale()
t = ts.now() # Current live time

# 4. Raw data ko process karke ek list mein convert karna
satellites_detected = []
for i in range(0, len(tle_data) - 2, 3):
    name = tle_data[i].strip()
    line1 = tle_data[i+1]
    line2 = tle_data[i+2]

    try:
        # Har single satellite ka mathematical model taiyar karna
        from skyfield.sgp4lib import EarthSatellite
        sat = EarthSatellite(line1, line2, name, ts)

        # Live coordinates calculate karna (Subpoint)
        geocentric = sat.at(t)
        subpoint = geocentric.subpoint()

        satellites_detected.append({
            "name": name,
            "lat": subpoint.latitude.degrees,
            "lon": subpoint.longitude.degrees,
            "alt": subpoint.elevation.km
        })
    except:
        continue

print(f"📊 Success: {len(satellites_detected)} Active Satellites are currently being tracked live in our system memory!")

# 5. Pehle 5 satellites ka live radar output check karna
print("\n--- SAMPLE RADAR STREAM (FIRST 5 OBJECTS) ---")
for s in satellites_detected[:5]:
    print(f"Object: {s['name']} | Lat: {s['lat']:.4f} | Lon: {s['lon']:.4f} | Altitude: {s['alt']:.2f} km")
print("---------------------------------------------")
```
